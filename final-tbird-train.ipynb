```python
BATCH_SIZE = 16
VAL_BATCH_SIZE = 32

import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline
from transformers import AutoModel
import datasets
import sklearn 
from sklearn.metrics import f1_score, accuracy_score, recall_score
from torch.utils.data import DataLoader

def get_model_tokenizer(model_name):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    return model, tokenizer

'''
Train above model on typical classification task
Two parts:
1. `load_data` returns a huggingface dataset
2. `train_model`
'''
def load_data():
    return datasets.load_dataset("glue", "sst2")

def load_data():
    return splitted_dataset

trainloader = DataLoader(load_data()['train'],
                      batch_size=BATCH_SIZE)
validation_loader = DataLoader(load_data()['validation'],
                      batch_size=VAL_BATCH_SIZE)
```

```python
for batch_num, batch in enumerate(validation_loader):
    print(type(
        batch['sentence']
    ))
    print(batch['sentence'])
    
    print(len(
        batch['label']
    ))
    print(batch['label'])
    
    if batch_num >= 32:
        break
```

```python
import gc
gc.collect()

import numpy as np
import json

BATCH_SIZE = 16

def train_model(model, tokenizer, train_loader, 
                val_loader=validation_loader,
                epochs=5, val=True):
    # Use training loop, don't use huggingface trainer

    # Train for 1 epoch
    # Use AdamW optimizer
    # Use CrossEntropyLoss
    # Use accuracy and f1 metric
    # Use batch size of 32
    # Use learning rate of 1e-5
    # Begin generation
    opt = torch.optim.AdamW(model.parameters(), 
                            lr=1e-5, 
                            weight_decay=1e-4)

    for epoch in range(epochs):
        # Train
        for batch_num, batch in enumerate(train_loader):
            # Tokenize
            # Forward pass
            # print(f'batch_num={batch_num}')
            sentences = batch['sentence']
            # sentences = [
            #     json.dumps(
            #         {'raw_text': sentence,
            #          'word_count': len(sentence.split())}
            #     ) for sentence in batch['sentence']
            # ]
            encodings = tokenizer(sentences, truncation=True, 
                                  padding=True, return_tensors='pt')
            encodings = encodings.to('cuda')
            # print(f'encodings of {batch_num} done')
            out = model(**encodings,  
                labels=batch["label"].to('cuda')) # convert int  to torch tensor
            # print(f'output of {batch_num} of {epoch} done')
            # Backward pass
            loss = out.loss
            loss.backward()
            opt.step()
            opt.zero_grad()
            # print metrics on validation set, f1 and accuracy
            # f1 and accuracy requires manual calculation 
            # or use something from sklearn, such as `sklearn.metrics`
            # get current batch_num
            # print(f'backward of {batch_num} of {epoch} done')
            print(f'batch:{batch_num} epoch:{epoch} loss:{loss.item()}')

            @torch.no_grad()
            def val_func():
                print("Batch num: ", batch_num)
                print("Loss: ", loss.item())
                # Validation
                y_true = []
                y_pred = []
                for idx, batch in enumerate(validation_loader):
                    encodings = tokenizer(batch["sentence"], 
                                          truncation=True, 
                                          padding=True,
                                          return_tensors='pt').to('cuda')
                    out = model(**encodings)
                    y_true.extend(batch["label"])
                    y_pred.extend(torch.argmax(out.logits, 
                                               dim=1).cpu())                
                y_pred = np.array(y_pred)
                f1 = f1_score(y_true, y_pred)
                acc = accuracy_score(y_true, y_pred)
                recall = recall_score(y_true, y_pred)

                print("Recall on validation: ", recall)
                print("F1 on validation: ", f1)
                print("Accuracy on validation: ", acc)
                
            if batch_num > 0 and batch_num % 10 == 0:
                val_func()

```

```python
model, tokenizer = get_model_tokenizer("microsoft/unixcoder-base")

model.to('cuda')

# Load data
dataset = load_data()
# training
train_model(model, tokenizer, 
            train_loader=trainloader,
           val_loader=validation_loader)
```

```python
def eval_func():
    with torch.no_grad():
        y_true = []
        y_pred = []
        for batch in validation_loader:
            encodings = tokenizer(batch["sentence"],
                                  truncation=True,
                                  padding=True,
                                  return_tensors='pt').to('cuda')
            out = model(**encodings)
            y_true.extend(batch["label"])
            y_pred.extend(torch.argmax(out.logits,
                                       dim=1).cpu())
        y_pred = np.array(y_pred)
        f1 = f1_score(y_true, y_pred)
        acc = accuracy_score(y_true, y_pred)
        recall = recall_score(y_true, y_pred)
        print("Accuracy on validation: ", acc)
        print("F1 on validation: ", f1)
        print("Recall on validation: ", recall)

    return acc, f1, recall
```
